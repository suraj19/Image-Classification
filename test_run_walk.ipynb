```python
import tensorflow 
import cv2
import numpy
#from PIL import Image
import matplotlib.pyplot as plt
```

```python
run_image = cv2.imread('C:/Users/Lenovo/Documents/GitHub/Image-Classification/download.jpg')
walk_image = cv2.imread('C:/Users/Lenovo/Documents/GitHub/Image-Classification/pexels.jpeg')
```

```python
plt.subplot(121)
plt.imshow(run_image)
plt.subplot(122)
plt.imshow(walk_image)
plt.show()
```

```python
print('shape of the run_image:{} and walk image:{}'.format(run_image.shape, walk_image.shape))
```

```python
dsize = (224, 224)
run_resized = cv2.resize(run_image, dsize)
walk_resized = cv2.resize(walk_image, dsize)
```

```python
plt.subplot(121)
plt.imshow(run_resized)
plt.subplot(122)
plt.imshow(walk_resized)
plt.show()
```

```python
run_resized = run_resized.reshape(1, 224, 224, 3)
walk_resized = walk_resized.reshape(1, 224, 224, 3)
```

```python
model = tensorflow.keras.models.load_model('C:/Users/Lenovo/Documents/GitHub/Image-Classification/run_walk_model.h5')
```

```python
model.summary()
```

```python
output = model.predict_classes(walk_resized)
if output == 0:
    print('The predicted image is: Run')
else:
    print('The predicted image is: walk')
```
